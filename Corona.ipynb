```python
# 지도에 표시 하기 위해 import
import folium
from folium import plugins
```

```python
# 지도에 표시하여 출력
def corona_place(place):
    place = corona(place)['Place']
    
    place_address = []
    place_lat = []
    place_lng = []
    points = []
    
    # 감염자 이동 경로의 주소 구하기
    for name in place:
        tmp = gmaps.geocode(name, language='ko')
        place_address.append(tmp[0].get("formatted_address"))
    
        tmp_loc = tmp[0].get("geometry")
    
        place_lat.append(tmp_loc['location']['lat'])
    
        place_lng.append(tmp_loc['location']['lng'])

    # 인천공항을 기준으로 지도가 출력된다
    m = folium.Map(location=[37.4601908, 126.4406957],
             zoom_start=10)
        
    for i in range(len(place_lat)):
        points.append([place_lat[i], place_lng[i]])
        
    # 지도에 마커를 추가 하고 마커들을 이어준다
    for index,lat in enumerate(place_lat):
        folium.Marker([lat, 
                       place_lng[index]],
                      popup=('감염자'.format(index)),
                      icon = folium.Icon(color='red',icon='plus')).add_to(m)
    folium.PolyLine(points, color='firebrick').add_to(m)
    
    return m
# 12번째 감염자의 이동 경로 출력       
corona_place(12)
```
